### 3. 构建注意力Decoder

解码器的输入包括循环网络最后的隐含状态$s_{i-1}$、最后输出$y_{i-1}$、所有编码器的所有输出$h_\ast$。  
1）这些输入，分别通过不同的层接受，$y_{t-1}$作为嵌入层的输入。


2）注意力层的函数a的输入为$s_{t-1}$和$h_j$，输出为$e_{tj}$，标准化处理后为$\alpha_{tj}$

3）向量$C_t$为编码器各输出的注意力加权平均。

（4）循环层f的输入为($s_{t-1}$,$y_{t-1}$,$c_t$),输出为内部隐含状态及$s_t$

（5）输出层g的输入为($y_{i-1}$,$s_i$,$c_i$)  ，输出为：$y_i$

（6）综合以上各步，得到Bahdanau注意力的解码器。
